```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import operator
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# import string
# import nltk
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

%matplotlib inline
```

```python
train = pd.read_csv('Resources/tweet_data.csv')
```

```python
train.head()
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment(sentence):
    score = analyser.polarity_scores(sentence)
    value = max(score.items(), key=operator.itemgetter(1))[0]
    returnValue = 0
    if value == "pos":
        returnValue = 1
    if value == "neg":
        returnValue = -1
    return returnValue
```

```python
train['stemmed_tweet'] = train['stemmed_tweet'].astype(str)
train['tidy_tweet'] = train['tidy_tweet'].astype(str)
```

```python
train['sentiment'] = train['tidy_tweet'].apply(lambda x: sentiment(x))
```

```python
train = train.drop(columns=['no_query'])
```
